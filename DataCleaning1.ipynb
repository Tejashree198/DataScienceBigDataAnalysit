```python
import pandas as pd  #To perform the operations on tabular dataset
import numpy as np #To perform mathematical operations. ex: Operations on Array
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df['marks'] = df['marks'].astype('int')
df['marks'].dtypes
```

```python
df['age'] = df['age'].astype('int')
df['age'].dtypes
```

```python
df["phd"] = df["phd"].replace('yes', 1)
df['phd'] = df['phd'].replace('no', 0)

print(df)
```

```python
df['phd'] = df['phd'].astype(int)
df['phd'].dtypes
```
